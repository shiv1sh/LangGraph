```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
load_dotenv()
```

```python
class BlogState(TypedDict):
    topic: str
    outline: str
    content: str

graph = StateGraph(BlogState)

model = ChatOpenAI(model = 'gpt-4o-mini')

def generate_outline(state: BlogState) -> BlogState:
    topic = state['topic']

    prompt = f'Generate a outline on this topic {topic}'

    outline = model.invoke(prompt)

    state['outline'] = outline

    return state

def generate_content(state : BlogState) -> BlogState:
    outline = state['outline']

    topic = state['topic']

    prompt = f'Generate a report following this outline and topic \n {outline} \n {topic}'

    content = model.invoke(prompt)

    state['content'] = content

    return state


```

```python
graph.add_node("outline_node", generate_outline)
graph.add_node("content_node", generate_content)

graph.add_edge(START, "outline_node")
graph.add_edge("outline_node", "content_node")
graph.add_edge("content_node", END)
```

```python
workflow = graph.compile()


response = workflow.invoke({"topic": "AI in India"})

print(response['content'])
```
