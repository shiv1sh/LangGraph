```python
from langgraph.graph import StateGraph, START, END
from langchain_openai import ChatOpenAI
from typing import TypedDict, Annotated
from pydantic import BaseModel, Field
from dotenv import load_dotenv
import operator
load_dotenv()
```

```python
class OutputStructure(BaseModel):
    output_score : int = Field(description = 'Score out of 10', ge=0 ,le= 10)
    feedback : str = Field(description = 'Detailed feedback for the essay')


model = ChatOpenAI(model = 'gpt-4o-mini')

structured_model = model.with_structured_output(OutputStructure)
```

```python
class UPSCState(TypedDict):
    essay : str
    language_feedback: str
    analysis_feedback : str
    clarity_feedback : str
    overall_feedback : str
    individual_score : Annotated[list[int], operator.add]
    avg_score: float
```

```python
def language_eval(state: UPSCState):
    essay = state['essay']
    prompt = f'Analyse the language quality and generate the feedback and score {essay}'
    response = structured_model.invoke(prompt)

    return {"language_feedback" : response.feedback, 'individual_score': [response.output_score]}

def analysis_eval(state: UPSCState):
    essay = state['essay']
    prompt = f'Analyse the analysis and generate the feedback and score {essay}'
    response = structured_model.invoke(prompt)

    return {"analysis_feedback" : response.feedback, 'individual_score': [response.output_score]}

def clarity_eval(state: UPSCState):
    essay = state['essay']
    prompt = f'Analyse the clarity and generate the feedback and score {essay}'
    response = structured_model.invoke(prompt)

    return {"clarity_feedback" : response.feedback, 'individual_score': [response.output_score]}

def overall_eval(state: UPSCState):
    
    prompt = f'Based on the feedbacks generate a summarized feedback /n {state['analysis_feedback']} \n {state['clarity_feedback']}  \n {state['language_feedback']}'
    overall_feedback = model.invoke(prompt)

    avg_score = sum(state['individual_score'])/ len(state['individual_score'])

    return {"overall_feedback" : overall_feedback, 'avg_score': avg_score}
```

```python
graph = StateGraph(UPSCState)
graph.add_node('language_node', language_eval)
graph.add_node('analysis_node', analysis_eval)
graph.add_node('clarity_node', clarity_eval)
graph.add_node('overall_node', overall_eval)

graph.add_edge(START, 'language_node')
graph.add_edge(START, 'analysis_node')
graph.add_edge(START, 'clarity_node')
graph.add_edge('language_node', 'overall_node')
graph.add_edge('analysis_node', 'overall_node')
graph.add_edge('clarity_node', 'overall_node')

graph.add_edge('overall_node', END)

workflow = graph.compile()
```
